### 2.2. Loading and Chunking Texts

We first need to load the documents. Here is an example:

```python
loader = WebBaseLoader(
    web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("post-content", "post-title", "post-header")
        )
    ),
)

documents = loader.load()
```

We will need to split the  text in chunks that are 'suitable' as retrieval units. Let's for starters define a chunk size of 128 and have no overlap between the chunks:  


```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=128, chunk_overlap=0)
splits = text_splitter.split_documents(documents)
print('Number of splits/chunks: ', str(len(splits)))
```

Ok, so it looks like we have now many splits (chunks) from one document. Here is how you can get the content:

```python
splits[39].page_content
```

Perfect. Now we have the splits and embeddings. Next, the embeddings need to be stored in a vector db.

### 2.3 Storing the Embeddings of Chunks in Vectorstores

After loading and chunking the data, we need to save the vector representations of the chunks in a vectorstore. We will use Qdrant here for simplicity. We load the splits (structured chunks) and the embeddings:

```python
vectorstore = Qdrant.from_documents(splits,
    base_embeddings,
    location=":memory:",  # Local mode with in-memory storage only
    collection_name="test",
)
retriever = vectorstore.as_retriever()
```

The nice thing is that the vector store also does the similarity searches for us:

```python
query = "What is Chain of Thought doing?"
docs = vectorstore.similarity_search_by_vector(base_embeddings.embed_query(query)) # will rank the splits
```

```python
docs
```

Looks good! We have an ordered list of documents that seem to relate to the question. That is what we need.

### 2.5 Testing the LLM in a LangChain Chain

Chains will be defined and discussed in Week 10. In short, they are convenient programmatic ways to deal with 'chains' of actions that involve LLMs. For example, a list of events like 'here is a city name. Plug that city name into prompt template, then generate a story about that city. Lastly, format the model output as a string' can be easily handled by LangChain's Chain framework. In this case, the Chain would consist of the prompt template, the LLM, and the String Formatter. The parameter (the city in this case) will be provided at run time by invocation of the Chain. Let's test that.

To use a Hugging Face model in a LangChain environment, we need to wrap the model into a LangChain pipeline object:

```python
mistral_llm_lc = HuggingFacePipeline(pipeline=mistral_pipe)
```

Next, we need to define a template and create a corresponding prompt template that can take any questiion

```python
test_llm_template = """[INST] Give me a two-sentence story about an {object}! [/INST]"""
test_llm_prompt_template = PromptTemplate(template=test_llm_template, input_variables=["object"])

```

Let's define a Chain, a static flow of actions that (usually) involve at least a definition of the variables used in the chain, one or more templates, LLM step(s) and potentially other actions. This would be a chain that declares the variable 'object' to be expected when the chain is invoked, then inserts it into the template, and passes this to our mistral model pipeline (wrapped as a LangChain object):    

```python
test_llm_chain_short = (
    {"object": RunnablePassthrough()}
    | test_llm_prompt_template
    | mistral_llm_lc
)
```

```python
test_llm_chain_short.invoke('apple')
```

Works too. We will use this notation moving forward.

### 2.6 Setting Up a Simple RAG Chain

For RAG, we will follow the same approach. Except... you will **later** need to change the chain to include the retrieval step.

We first do a simple test: create a RAG template that takes a question and a pre-defined context as input, and generates the answer based on the provided context:

```python
rag_template = """[INST] Answer the question based only on the following context:
{context}

Question: {question}
[/INST]
"""
rag_prompt_template = ChatPromptTemplate.from_template(rag_template)

base_rag_chain =(
    {"context": RunnablePassthrough(),
     "question": RunnablePassthrough()}
    | rag_prompt_template
    | mistral_llm_lc
    | output_parser
)

predefined_context = "Germany has won the World Cup 4 times."
question = "How many times did Germany win the world cup?"

resp = base_rag_chain.invoke({'context': predefined_context,
                           'question': question})
print(resp)
```

That's great. But of course, the context needs to be created in an earlier retrieval step. More precisely, the documents will be first retrieved as a list, and then they will need to be formatted into one string to pass to the LLM in the context window.

Here is a simple formatting function that can be hooked into the chain, which combines a list of chunks into one string:



```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

So how could we build a simple chain? Let's first just get the retrieval done and the formatted retrieved data and the question inserted into the prompt template:

```python
rag_template = """Here is a context:\n{context} \n\nand here is a question: \n{question}"""

rag_prompt = ChatPromptTemplate.from_template(rag_template)

rag_chain = (
    {"context": retriever | format_docs,
     "question": RunnablePassthrough()}
    | rag_prompt


)
```

```python
output = rag_chain.invoke('What is Chain of Thought?')
```

Ok... with some formatting... this looks good:

```python
print(output.messages[0].content)
```

Let's complete the RAG Chain:

```python
output_parser = StrOutputParser()

rag_template = """[INST]Please answer the question below only based on the context information provided.\n\nHere is a context:\n{context} \n\nHere is a question: \n{question}.[/INST]"""
rag_prompt = ChatPromptTemplate.from_template(rag_template)

rag_chain = (
    {"context": retriever | format_docs,
     "question": RunnablePassthrough()}
    | rag_prompt
    | mistral_llm_lc

)

```

```python
rag_chain.invoke('What is Chain of Thought?')
```

What about the Cohere models?

```python
cohere_rag_chain = (
    {"context": retriever | format_docs,
     "question": RunnablePassthrough()}
    | rag_prompt
    | cohere_chat_model
    | output_parser
)
```

```python
cohere_rag_chain.invoke('What is Chain of Thought?')
```

Works too! Time to build the real thing and do experimentation.

First we'll grab some papers from ArXiv.  We'll grab the pdf files and get all of the pages as separate documents.

```python
arxiv_numbers = ('2005.11401', '2104.07567', '2104.09864', '2105.03011', '2106.09685', '2203.02155', '2211.09260', '2211.12561',
                 '2212.09741', '2305.14314', '2305.18290', '2306.15595', '2309.08872', '2309.15217', '2310.06825', '2310.11511',
                 '2311.08377', '2312.05708', '2401.06532', '2401.17268', '2402.01306', '2402.19473', '2406.04744',
                 '2312.10997', '2410.12812', '2410.15944', '2404.00657',
                 )
```

```python
# Document formatting function
def format_docs(docs):
    """Format retrieved documents into a single context string"""
    return "\n\n".join(f"Source: {doc.page_content}" for doc in docs)

# Create retriever from the vector store
# The retriever converts the vector store into a searchable component for RAG chains
# search_kwargs={"k": 4} means it will return the top 4 most similar document chunks
# when searching. This is the "retrieval" part of RAG - it finds relevant context
# for the user's question by comparing question embeddings to document embeddings.
retriever = qdrant_vectorstore.as_retriever(search_kwargs={"k": 4})

# Output parser for clean string responses
# Different LLM providers return responses in different formats (e.g., AIMessage objects)
# StrOutputParser() extracts just the text content to ensure consistent, clean string
# outputs regardless of which LLM (Mistral, Cohere, etc.) is being used in the chain.
output_parser = StrOutputParser()
```

```python
def safe_float(value):
    """Convert value to float, return None if invalid"""
    # print(f"DEBUG safe_float: input={value}, type={type(value)}")
    if value is None:
        # print("DEBUG safe_float: returning None (input was None)")
        return None
    try:
        # Convert to Python float first, then check if it's valid
        float_val = float(value)
        if np.isnan(float_val):
            # print(f"DEBUG safe_float: returning None (NaN detected)")
            return None
        else:
            # print(f"DEBUG safe_float: returning {float_val}")
            return float_val
    except (ValueError, TypeError) as e:
        print(f"DEBUG safe_float: exception {e}, returning None")
        return None
```

```python
def get_calibrated_weights():
    """
    Weights based on discrimination power after normalizing for natural score ranges
    """
    # Step 1: Natural ranges observed in calibration
    normalization_params = {
        'semantic_similarity': {'min': 0.168, 'max': 1.000},  # Range: 0.832
        'rouge_l': {'min': 0.000, 'max': 1.000},              # Range: 1.000
        'bleu': {'min': 0.000, 'max': 1.000},                 # Range: 1.000
        'bert_score_f1': {'min': 0.798, 'max': 1.000}         # Range: 0.202 (narrow!)
    }

    # Step 2: Discrimination power after normalization
    # When we normalize BERTScore from 0.798-1.000 to 0-1, its discrimination of 0.026
    # becomes 0.026/0.202 = 0.129 in normalized space
    # When we normalize others, discrimination stays roughly the same since they use full range

    normalized_discrimination = {}
    raw_discrimination = {
        'semantic_similarity': 0.150,  # 0.150/0.832 = 0.180 normalized
        'bleu': 0.048,                 # 0.048/1.000 = 0.048 normalized
        'bert_score_f1': 0.026,        # 0.026/0.202 = 0.129 normalized
        'rouge_l': -0.185              # -0.185/1.000 = -0.185 normalized
    }

    for metric, disc_score in raw_discrimination.items():
        if metric in normalization_params:
            range_size = normalization_params[metric]['max'] - normalization_params[metric]['min']
            if range_size > 0:
                normalized_discrimination[metric] = disc_score / range_size
            else:
                normalized_discrimination[metric] = 0.0
        else:
            normalized_discrimination[metric] = 0.0

    # Convert to weights based on normalized discrimination
    raw_weights = {}
    for metric, norm_disc in normalized_discrimination.items():
        if norm_disc > 0:  # Correct direction
            raw_weights[metric] = norm_disc
        elif norm_disc < 0:  # Wrong direction - penalize
            raw_weights[metric] = abs(norm_disc) * 0.1  # 10% penalty
        else:
            raw_weights[metric] = 0.0

    # Add disabled metrics
    raw_weights['faithfulness'] = 0.0
    raw_weights['answer_relevancy'] = 0.0
    raw_weights['context_precision'] = 0.0

    # Normalize weights to sum to 1.0
    total = sum(raw_weights.values())
    return {k: v/total for k, v in raw_weights.items()} if total > 0 else {'semantic_similarity': 1.0}
```

```python
def calibrate_all_metrics():
    """Test all metrics on carefully designed examples to understand score ranges"""

    # Categorize test cases by EXPECTED performance
    test_cases = {
        'should_score_high': [
            # These SHOULD get high scores (same/similar meaning)
            ("What is your age?", "How old are you?", "same_meaning_different_words"),
            ("The car is red", "The car is crimson", "synonyms"),
            ("Large language models use attention", "Big neural networks employ attention mechanisms", "good_paraphrase"),
            ("The algorithm performs classification", "The method does categorization tasks", "good_paraphrase"),
            ("Training requires large datasets", "Learning needs lots of data", "good_paraphrase"),
            ("What is a transformer model?", "What is a transformer model?", "perfect_match"),
        ],

        'should_score_medium': [
            # These might get medium scores (partial overlap, related but incomplete)
            ("Transformers use self-attention mechanisms", "BERT uses attention", "partial_overlap"),
            ("GPT-3 has 175 billion parameters", "GPT models have many parameters", "partial_overlap"),
            ("BERT uses bidirectional encoder representations", "BERT reads text in both directions", "technical_vs_simple"),
            ("Yes", "Absolutely, that is correct and I completely agree", "length_mismatch_same_meaning"),
        ],

        'should_score_low': [
            # These SHOULD get low scores (different meaning, wrong info, unrelated)
            ("The model is accurate", "The model is inaccurate", "opposite_meaning"),
            ("Turn left at the corner", "Turn right at the corner", "opposite_direction"),
            ("GPT-3 was released by OpenAI", "BERT was developed by Google", "different_facts"),
            ("BERT has 12 layers", "BERT has 24 layers", "wrong_factual_details"),
            ("What defines a large language model?", "I love eating pizza on weekends", "completely_unrelated"),
            ("BERT is a neural network", "BERT is a pizza restaurant", "pure_hallucination"),
        ]
    }

    print("=== METRIC CALIBRATION BY EXPECTED PERFORMANCE ===\n")

    # Process each expected performance category
    calibration_results_by_expectation = {}

    for expectation, cases in test_cases.items():
        print(f"{'='*60}")
        print(f"TESTING: {expectation.upper().replace('_', ' ')}")
        print('='*60)

        test_case_results = []
        # print(f"DEBUG: Starting {expectation}, test_case_results is empty list: {len(test_case_results)}")

        for text1, text2, description in cases:
            # print(f"\nDEBUG: About to call evaluate_against_gold for {description}")
            metric_scores = evaluate_against_gold(text1, text2)

            # print(f"DEBUG: Returned metric_scores for {description}:")
            for metric, score in metric_scores.items():
                print(f"  DEBUG: {metric} = {score}")

            test_case_data = {
                'text1': text1,
                'text2': text2,
                'description': description,
                'scores': metric_scores
            }

            # print(f"DEBUG: About to append to test_case_results. Current length: {len(test_case_results)}")
            test_case_results.append(test_case_data)
            # print(f"DEBUG: After append, test_case_results length: {len(test_case_results)}")

            print(f"\n{description}:")
            print(f"  '{text1}' vs '{text2}'")
            for metric, score in metric_scores.items():
                if score is not None:
                    print(f"    {metric}: {score:.3f}")

        # print(f"DEBUG: Finished {expectation}, storing {len(test_case_results)} results")
        calibration_results_by_expectation[expectation] = test_case_results

        # Calculate averages for this expectation category
        print(f"\nAVERAGE SCORES for {expectation}:")
        if test_case_results:
            for metric in test_case_results[0]['scores'].keys():
                avg_score = np.mean([r['scores'][metric] for r in test_case_results if r['scores'][metric] is not None])
                print(f"  {metric}: {avg_score:.3f}")

    # Generate interpretation guide based on expected performance
    print(f"\n{'='*80}")
    print("METRIC-SPECIFIC INTERPRETATION GUIDE")
    print('='*80)

    for metric in calibration_results_by_expectation['should_score_high'][0]['scores'].keys():
        print(f"\n{metric.upper()}:")

        # Calculate ranges for each expectation
        high_scores = [r['scores'][metric] for r in calibration_results_by_expectation['should_score_high']
                      if r['scores'][metric] is not None]
        medium_scores = [r['scores'][metric] for r in calibration_results_by_expectation['should_score_medium']
                        if r['scores'][metric] is not None]
        low_scores = [r['scores'][metric] for r in calibration_results_by_expectation['should_score_low']
                     if r['scores'][metric] is not None]

        if high_scores:
            high_min, high_max = min(high_scores), max(high_scores)
            high_avg = np.mean(high_scores)
            print(f"  Good matches (should be high): {high_min:.3f} - {high_max:.3f} (avg: {high_avg:.3f})")

        if medium_scores:
            med_min, med_max = min(medium_scores), max(medium_scores)
            med_avg = np.mean(medium_scores)
            print(f"  Partial matches (should be medium): {med_min:.3f} - {med_max:.3f} (avg: {med_avg:.3f})")

        if low_scores:
            low_min, low_max = min(low_scores), max(low_scores)
            low_avg = np.mean(low_scores)
            print(f"  Poor matches (should be low): {low_min:.3f} - {low_max:.3f} (avg: {low_avg:.3f})")

        # Discrimination analysis
        if high_scores and low_scores:
            separation = np.mean(high_scores) - np.mean(low_scores)
            overlap = max(low_scores) >= min(high_scores)
            print(f"  Discrimination: {separation:.3f} separation, {'OVERLAP' if overlap else 'CLEAN SEPARATION'}")

    # Practical thresholds
    print(f"\n{'='*80}")
    print("SUGGESTED PRACTICAL THRESHOLDS")
    print('='*80)

    for metric in calibration_results_by_expectation['should_score_high'][0]['scores'].keys():
        high_scores = [r['scores'][metric] for r in calibration_results_by_expectation['should_score_high']
                      if r['scores'][metric] is not None]
        low_scores = [r['scores'][metric] for r in calibration_results_by_expectation['should_score_low']
                     if r['scores'][metric] is not None]

        if high_scores and low_scores:
            # Suggest threshold as midpoint between averages
            high_avg = np.mean(high_scores)
            low_avg = np.mean(low_scores)
            threshold = (high_avg + low_avg) / 2

            # Calculate accuracy at this threshold
            correct_high = sum(1 for s in high_scores if s >= threshold)
            correct_low = sum(1 for s in low_scores if s < threshold)
            accuracy = (correct_high + correct_low) / (len(high_scores) + len(low_scores))

            print(f"\n{metric.upper()}:")
            print(f"  Suggested threshold: {threshold:.3f}")
            print(f"  Accuracy at threshold: {accuracy:.1%}")
            print(f"  Good answers ≥ {threshold:.3f}, Poor answers < {threshold:.3f}")

    return calibration_results_by_expectation
```

```python
def test_rag_system(chain, persona_type, retriever_instance, question_subset=None):
    """
    Test RAG chain - now with comprehensive metrics
    """
    results = []
    questions_to_test = question_subset if question_subset else validation_questions_answers.keys()

    for i, q_id in enumerate(questions_to_test):
        question_data = validation_questions_answers[q_id]
        question = question_data['question']

        gold_key = f'gold_answer_{persona_type}'
        if gold_key not in question_data:
            continue

        gold_answer = question_data[gold_key]

        try:
            print(f"Processing Q{q_id} ({i+1}/{len(list(questions_to_test))}")

            # Get retrieved contexts
            retrieved_docs = retriever.invoke(question)
            retrieved_contexts = [doc.page_content for doc in retrieved_docs]

            # Generate response
            try:
                generated_response = chain.invoke(question)
            except Exception as e:
                if "rate limit" in str(e).lower():
                    print(f"Rate limit hit for Q{q_id}, waiting 60 seconds...")
                    time.sleep(60)
                    try:
                        generated_response = chain.invoke(question)
                    except Exception as e2:
                        print(f"Still failing after retry: {e2}")
                        continue
                else:
                    raise e

            # Comprehensive evaluation
            scores = evaluate_against_gold(
                generated_response,
                gold_answer,
                question=question,
                retrieved_contexts=retrieved_contexts
            )

            # Calculate combined score for this result - ONLY using available metrics
            weights = get_calibrated_weights()
            weighted_score = 0.0
            total_weight = 0.0

            for metric, weight in weights.items():
                if weight > 0 and metric in scores and scores[metric] is not None:
                    if isinstance(scores[metric], (int, float)) and not np.isnan(scores[metric]):
                        weighted_score += scores[metric] * weight
                        total_weight += weight

            # Normalize by actual total weight used - this ensures fairness
            combined_score = weighted_score / total_weight if total_weight > 0 else 0.0

            results.append({
                'question_id': q_id,
                'question': question,
                'generated': generated_response,
                'gold': gold_answer,
                'retrieved_contexts': retrieved_contexts,
                'doc_sources': [doc.metadata.get('doc_source', 'Unknown') for doc in retrieved_docs],
                'combined_score': combined_score,
                **scores  # Unpack all metric scores
            })

            # Rate limiting
            if 'cohere' in str(type(chain)).lower():
                time.sleep(10)

        except Exception as e:
            print(f"Error processing question {q_id}: {e}")

    return results
```

**Generated Engineering Response**:
A **Large Language Model (LLM)** in the context of natural language processing (NLP) is defined by several key characteristics, including its architecture, scale, training methodology, and capabilities. Below is a comprehensive technical breakdown of what defines an LLM:

---

### **1. Architectural Foundation**
LLMs are typically based on **Transformer architectures**, specifically the **encoder-decoder** or **decoder-only** variants. The Transformer model, introduced in the paper *"Attention is All You Need"* (Vaswani et al., 2017), is the backbone of modern LLMs. Key components include:

-   **Self-Attention Mechanism**: Allows the model to weigh the importance of different words in a sequence, enabling context-aware representations.
-   **Multi-Head Attention**: Extends self-attention by computing multiple attention distributions in parallel, capturing diverse contextual relationships.
-   **Feed-Forward Neural Networks**: Applied to each position separately after attention layers to introduce non-linear transformations.
-   **Layer Normalization and Residual Connections**: Stabilize training and improve gradient flow in deep networks.

**Example**: GPT (Generative Pre-trained Transformer) models use a decoder-only architecture, while BERT (Bidirectional Encoder Representations from Transformers) uses an encoder-only architecture.

---

### **2. Scale: Size and Complexity**
The term "large" in LLMs refers to their massive scale, both in terms of **model parameters** and **training data**. Typical LLMs have:

-   **Parameters**: Ranging from hundreds of millions to hundreds of billions (e.g., GPT-3 has 175 billion parameters, while smaller models like GPT-2 have 1.5 billion).
-   **Training Data**: Trained on vast corpora of text data, often exceeding 100 GB of text (e.g., GPT-3 was trained on 570 GB of text from diverse sources like books, websites, and articles).

**Example**: GPT-3's scale enables it to generate human-like text across a wide range of tasks, whereas smaller models may struggle with complexity or generalization.

---

### **3. Training Methodology**
LLMs are trained using **unsupervised learning** on large text corpora. The training process involves:

-   **Pre-training**: The model learns to predict the next word in a sequence (causal language modeling) or mask-fill missing words (masked language modeling). This phase is computationally intensive and requires significant resources.
-   **Fine-tuning**: Optionally, the model is fine-tuned on task-specific datasets to adapt to particular NLP tasks (e.g., sentiment analysis, question answering).
-   **Objective Functions**: Typically, cross-entropy loss is used to maximize the likelihood of predicting the correct next word.

## 1. Computational Efficiency During Training
LoRA reduces the computational overhead of fine-tuning transformer models by introducing low-rank updates to a subset of weight matrices instead of updating the entire model. Here’s how it achieves efficiency:

#### **a. Reduced Parameter Updates**
- **Methodology**: LoRA approximates the update to a weight matrix $W$ (e.g., in self-attention layers) by decomposing it into the product of two low-rank matrices $A$ and $B$:
  $$W' = W + BA$$
  where $B \in \mathbb{R}^{d \times r}$ and $A \in \mathbb{R}^{r \times k}$, and $r$ is the rank of the decomposition (typically much smaller than $d$ and $k$).
- **Efficiency**: Instead of updating all $d \times k$ parameters in $W$, only $(d + k) \times r$ parameters in $A$ and $B$ are trained. For example, in a transformer with $d = 768$ and $r = 8$, the number of trainable parameters is reduced by a factor of $\frac{768 \times 768}{(768 + 768) \times 8} = 72$.

#### **b. Memory Efficiency**
- **Methodology**: During training, only the low-rank matrices $A$ and $B$ are stored in memory, while the original pre-trained weights $W$ remain frozen.
- **Efficiency**: This significantly reduces GPU memory usage compared to full fine-tuning, enabling training on hardware with limited resources. For instance, fine-tuning a 6B parameter model with LoRA (rank $r = 8$) requires only ~0.1% of the memory needed for full fine-tuning.

#### **c. Faster Convergence**
- **Methodology**: LoRA focuses on updating only the most critical parameters, which often leads to faster convergence during training.
- **Efficiency**: Fewer training iterations are required to achieve comparable performance to full fine-tuning, reducing overall training time.

---
## 2. Computational Efficiency During Deployment
LoRA’s design enables efficient deployment by minimizing the storage and computational requirements for serving adapted models.

#### **a. Small Model Footprint**
- **Methodology**: Only the low-rank adaptation matrices $A$ and $B$ need to be stored and deployed, while the base pre-trained model remains shared across tasks.
- **Efficiency**: For example, deploying 10 task-specific LoRA modules for a 6B parameter model requires storing only ~10 MB of additional parameters (assuming $r = 8$), compared to 24 GB for 10 fully fine-tuned models.

#### **b. Dynamic Model Switching**
- **Methodology**: At inference time, the adapted model is constructed by adding the low-rank updates to the frozen pre-trained weights:
  $$W' = W + BA$$
  This allows for dynamic switching between tasks by simply swapping the LoRA weights $A$ and $B$.
- **Efficiency**: Task switching incurs minimal overhead, as only the small LoRA matrices need to be loaded, rather than an entire model. For example, switching between two tasks in a real-time application can be done in milliseconds.

#### **c. Reduced Inference Cost**
- **Methodology**: Since the base model remains unchanged, inference computations are shared across tasks, and only the low-rank updates are applied during forward passes.
- **Efficiency**: This reduces the computational cost of inference, especially in scenarios with multiple task-specific adaptations.
